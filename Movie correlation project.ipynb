```python
# import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create 

# read in the data 

df = pd.read_csv('Downloads/movies (1).csv')
```

```python
 # Lets look at the data
    
df.head ()
```

```python
# Is there any missing data?

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
    df.isnull().sum()
    
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
    
# There isn't. 
```

```python
# Data types of the data

df.dtypes
```

```python
# Data types of the data

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# creating the correct year column

df = df.drop('correctyear', 1)
```

```python
# creating the correct year column

df['correct_year'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates if they exist

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Scatterplot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross')

plt.ylabel('Budget')


plt.show()
```

```python
df.head(10)
```

```python
# plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color":"red"})
```

```python
# Looking at the correlation matrix

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix')

plt.show()
```

```python
# Looking at the company and converting it into a numeric representation

df.head()
```

```python
df_numberised = df

for col_name in df_numberised.columns:
    if(df_numberised[col_name].dtype == 'object'):
        df_numberised[col_name] = df_numberised[col_name].astype('category')
        df_numberised[col_name] = df_numberised[col_name].cat.codes
        
df_numberised.head()
```

```python
df.head()
```

```python

correlation_matrix = df_numberised.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix')

plt.show()
```

```python
df_numberised.corr()
```

```python
correlation_mat = df_numberised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
